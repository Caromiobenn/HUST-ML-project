```python
import numpy as np # 数组常用库
import pandas as pd # 读入csv常用库
import matplotlib.pyplot as plt # 画图常用库
```

```python
#加载数据
def load_data( train_row):
    train = pd.read_csv("train.csv")
    print(train.shape)
    X_train = train.values[0:train_row,1:] # 取下标为1-784的列（pixel0 - pixel783）
    y_train = train.values[0:train_row,0] # 取下标为0的列 (label)
    Pred_test = pd.read_csv("test.csv").values 
    return X_train, y_train, Pred_test
train_row=5000
#train_row = 42000 
Origin_X_train, Origin_y_train, Origin_X_test = load_data(train_row)
```

```python
from sklearn.model_selection import train_test_split
#划分训练集与验证集
X_train,X_vali, y_train, y_vali = train_test_split(Origin_X_train,Origin_y_train,test_size = 0.2,random_state = 0)
```

```python
import time
from knn_implementation import KNNClassifier
```

```python
#在小规模下确定K取值
from sklearn.metrics import accuracy_score
k_range = range(3, 8)
scores = []
for k in k_range:
    print("k = " + str(k) + " begin ")
    start = time.time()
    knn = KNNClassifier(k)
    model = knn.fit(X_train,y_train)
    y_pred=model.evalue(X_vali,y_vali)
    accuracy = accuracy_score(y_vali,y_pred)   
    scores.append(accuracy)
    end = time.time()
    print("Complete time: " + str(end-start) + " Secs.")
```

```python
#展示准确率随K值变化情况
print (scores)
plt.plot(k_range,scores)
plt.xlabel('Value of K')
plt.ylabel('Testing accuracy')
plt.xticks(np.arange(3, 8, 1))
plt.show()
```

```python
#确定K=3后，对全体数据作训练和验证
k=3
knn = KNNClassifier(k)
model = knn.fit(X_train,y_train)
model.evalue(X_vali,y_vali)
```

```python
#对测试集，就算标签，并且写到submission.csv中
knn = KNNClassifier(3)
model = knn.fit(X_train,y_train)
y_pred=model.evalue(Origin_X_test,np.ones(28000))
print(len(y_pred))

pd.DataFrame({"ImageId": list(range(1,len(y_pred)+1)),"Label": y_pred}).to_csv('submission.csv', index=False,header=True)
```
